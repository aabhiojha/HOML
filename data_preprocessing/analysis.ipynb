### Removing urls

```python
import re
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

```python
remove_url(df['Review'][3715])
```

# Removing handeling with string library/module

```python
import string
```

```python
symbols=string.punctuation
symbols
```

```python
def remove_punct(text):
    for char in symbols:
        text=text.replace(char,'')
    return text
```

```python
text = "Damnn!! so many, punctuations!.@$%"
```

```python
remove_punct(text)
```

## Stemmer and Lemmatization
- Stemmer is not that accurate
- Lemmatization provides with accurate root words

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
text = "run running runs runningly"
```

```python
stem_words(text)
```
